```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-25,25,11)
```

```python
df = pd.read_csv('01Mass', header=None, delimiter=" ")
last = np.linspace(31, 351, 11)
evac_times1 = np.zeros((11,5))

for i in range(0,11):
    evac_times1[i]=df.loc[last[i]]
    
#evac_times[:,0]
#np.mean(evac_times, axis=0)

ave_collisions = np.zeros((11,1))

for i in range(0,11):
    lower = 32*i
    upper = (32*i)+31
    i_collisions = df.loc[lower:upper][3]
    ave_collisions[i]=np.mean(i_collisions, axis = 0)

y1 = (evac_times1[:,0]/evac_times1[5,0])-1
plt.scatter(x,y1)
plt.axvline(0, c='black', ls='--') 
plt.axhline(0, c='black', ls='--')
plt.show
```
